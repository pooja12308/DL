# Diabeties Risk Prediction(Binary Classification)
Neural Network architeture:
2i/ps->3 hidden neurons(sigmoid)->1o/p(sigmoid)

o/p=1->High risk

o/p=0->Low risk

```python
import numpy as np
np.random.seed(0)
X=np.array([
    [120,85,150,95], #Glucose
    [25,30,35,22]    #BMI
])
```

```python
#Normalize Values
X=X/200
```

```python
#Output:1-High Risk, 0-Low Risk
y=np.array([[1,0,1,0]])
```

```python
print("X shape:",X.shape)
print("y shape:",y.shape)
```

```python
def sigmoid(Z):
    return 1/(1+np.exp(-Z))
```

# Initialize Parameters

```python
#Hidden Layer(3neurons,2inputs)
W1=np.random.rand(3,2)*0.1
b1=np.zeros((3,1))

#Output Layer(1neuron,3inputs)
W2=np.random.rand(1,3)*0.1
b2=np.zeros((1,1))
```

```python
# Hidden Layer
z1=np.dot(W1,X)+b1
a1=sigmoid(z1)

# Output Layer
z2=np.dot(W2,a1)+b2
a2=sigmoid(z2)

print("Predicted Output:",a2)

```

# Loss Function
Binary Cross Entropy

```python
m=y.shape[1]
loss=-np.sum(y*np.log(a2+1e-8)+(1-y)*np.log(1-a2+1e-8))/m
print("Loss:",loss)
```

```python
#Back propagation
dz2=a2-y
dW2=np.dot(dz2,a1.T)/m
db2=np.sum(dz2,axis=1,keepdims=True)/m
da1=np.dot(W2.T,dz2)
dz1=da1*a1*(1-a1)
dW1=np.dot(dz1,X.T)/m
db1=np.sum(dz1,axis=1,keepdims=True)/m
```

```python
learning_rate=0.5
W2-=learning_rate*dW2
b2-=learning_rate*db2

W1-=learning_rate*dW1
b1-=learning_rate*db1
```

```python
epoch=5000
for i in range(epoch):
    # Forward pass
    z1=np.dot(W1,X)+b1
    a1=sigmoid(z1)
    z2=np.dot(W2,a1)+b2
    a2=sigmoid(z2)
    
    # Compute loss
    m=y.shape[1]
    loss=-np.sum(y*np.log(a2+1e-8)+(1-y)*np.log(1-a2+1e-8))/m
    
    # Backward pass
    dz2=a2-y
    dW2=np.dot(dz2,a1.T)/m
    db2=np.sum(dz2,axis=1,keepdims=True)/m
    da1=np.dot(W2.T,dz2)
    dz1=da1*a1*(1-a1)
    dW1=np.dot(dz1,X.T)/m
    db1=np.sum(dz1,axis=1,keepdims=True)/m
    
    # Update weights and biases
    W2-=learning_rate*dW2
    b2-=learning_rate*db2
    W1-=learning_rate*dW1
    b1-=learning_rate*db1  
    if i%100==0:
        print(f"Epoch {i}, Loss: {loss:.4f}")
         
```
