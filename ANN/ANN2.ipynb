```python
import numpy as np
#XOR 
X=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,1,1,0])
X,y
```

```python
w1,w2,w3,w4,w5,w6=1,1,-1,-1,1,1
b1=-0.5
b2=1.5
b3=-1.5
epochs=1
```

```python
def step_function(X):
    return np.where(X>=0,1,0)
```

```python
for x1, x2 in X:
    h1 = step_function(w1*x1 + w2*x2 + b1)
    h2 = step_function(w3*x1 + w4*x2 + b2)
    output = step_function(w5*h1 + w6*h2 + b3)
    print(x1, x2, "→", output)
```

# Another Method

```python
def forward_propagation(x1,x2):
    h1=w1*x1 + w2*x2 + b1
    h2=w3*x1 + w4*x2 + b2
    a_h1=step_function(h1)
    a_h2=step_function(h2)
    yo=w5*a_h1 + w6*a_h2 + b3
    a_y=step_function(yo)
    return a_y
```

# Test Forward Propagation For All XOR Inputs

```python
print('XOR results using ANN forward')
for x1, x2 in X:
    output=forward_propagation(x1,x2)
    print(x1, x2, "→", output)
```

```python
#Plotting the decision boundary(Now we get a straight line)
import matplotlib.pyplot as plt
x1=np.array([0,0,1,1])
x2=np.array([0,1,0,1])
y=np.array([0,1,1,0])
plt.scatter(x1,x2,c=y)
x1=np.linspace(-0.5,1.5,100)
x2=-w1*x1/b1
plt.plot(x1,x2,'r')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Decision Boundary for XOR using ANN')
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
plt.show()
```
