# First Example

```python
import numpy as np
#Training Data
X=np.array([1,2,3,4])
y=np.array([0,0,1,1])
X,y
```

# Define weights, Bias, Epochs & Learning Rate

```python
w=0.0
b=0.0
learning_rate=1
epochs=5

print("Initial Weight:",w)
print("Initial Bias:",b)
```

# Define Activation Function

```python
def step_function(z):
    return np.where(z >= 0, 1, 0)
```

# Training Loop

```python
for epoch in range(epochs):
    print(f"\nEpoch:{epoch+1}")
    for i in range(len(X)):
        #1.Compute Weighted Sum
        z=w*X[i]+b
        #2.Apply Activation Function
        y_pred=step_function(z)
        #3.Compute Error
        error=y[i]-y_pred
        #4.Update Weight
        w=w+learning_rate*error*X[i]
        #5.Update Bias
        b=b+learning_rate*error
        print(f"Input:{X[i]}\tPredicted:{y_pred}\tActual:{y[i]}\tError:{error}\tUpdated Weight:{w}\tUpdated Bias:{b}")
         
```

# Final Model Parameters

```python
print("\nFinal Weight:",w)
print("Final Bias:",b)
```

# Train The Model

```python
test_value=3.01
z=w*test_value+b
pred=step_function(z)
print(f"\nTest Input:{test_value}\tPredicted Output:{pred}")
```

# Plotting Decision Boundary

```python
import matplotlib.pyplot as plt
#Plotting the decision boundary
X=np.array([1,2,3,4])
y=np.array([0,0,1,1])
plt.scatter(X,y,color='blue',label='Data Points')
x_values=np.linspace(0,5,100)
y_values=step_function(w*x_values+b)
plt.plot(x_values,y_values,color='red',label='Decision Boundary')
plt.xlabel('Input (X)')
plt.ylabel('Output')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.grid()
plt.show()
```
