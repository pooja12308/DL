# For OR Operations


```python
import numpy as np

# OR dataset
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([0,1,1,1])

# Initialize randomly
w = np.random.randn(2)
b = np.random.randn()

learning_rate = 0.1
epochs = 10

def step_function(z):
    return np.where(z >= 0, 1, 0)

# Training
for epoch in range(epochs):
    for i in range(len(X)):
        z = np.dot(X[i], w) + b
        y_pred = step_function(z)
        
        error = y[i] - y_pred
        
        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error

print("Final Weights:", w)
print("Final Bias:", b)

# Predictions
print("Predictions:", step_function(np.dot(X, w) + b))
```

# For XOR

```python
import numpy as np

X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([0,1,1,0])  # XOR

w = np.random.randn(2)
b = np.random.randn()

learning_rate = 0.1
epochs = 20

def step_function(z):
    return np.where(z >= 0, 1, 0)

for epoch in range(epochs):
    total_error = 0
    
    for i in range(len(X)):
        z = np.dot(X[i], w) + b
        y_pred = step_function(z)
        error = y[i] - y_pred
        
        total_error += abs(error)
        
        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error
        
    print(f"Epoch {epoch+1}, Total Error: {total_error}")
```
