# For AND Operations (Two Inputs & One Output)

# Prepare Training Data

```python
import numpy as np
X=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
y=np.array([0,0,0,1])
X,y

```

# Define Weights, Bias, Epochs & Learning Rate

```python
w=np.zeros(2)
b=0
learning_rate=0.1
epochs=10

print("Initial Weight:",w)
print("Initial Bias:",b)
```

# Define Activation Function

```python
def step_function(z):
    return np.where(z >=0,1,0)
```

# Training Loop

```python
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    
    for i in range(len(X)):
        z = np.dot(X[i], w) + b
        y_pred = step_function(z)
        
        error = y[i] - y_pred
        
        # Update rule
        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error
        
        print(f"Input: {X[i]}, Pred: {y_pred}, Error: {error}")

print("\nFinal Weights:", w)
print("Final Bias:", b)
```

# Final Model Parameters

```python
print("\nFinal Weight:",w)
print("Final Bias:",b)
```

# Train the Model


```python
print("\n Testing AND Gate:")
for i in range(len(X)):
    z=np.dot(w,X[i])+b
    prediction=step_function(z)
    print("input:",X[i],"Prediction:",prediction)
```

# Plotting The Decision Boundary

```python
import matplotlib.pyplot as plt
#Plotting the decision boundary
x_values=np.array([0,5])
y_values=-(w[0]*x_values + b)/w[1]
plt.plot(x_values,y_values,label="Decision Boundary")
plt.scatter(X[:,0],X[:,1],c=y,label="Data Points")
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.title("Perceptron Decision Boundary")
plt.legend()
plt.grid()
plt.show()


```
