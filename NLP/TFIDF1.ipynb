```python
text=[
    "I love coding",
    "I love python",
    "Python is powerful"
]
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()

X=vectorizer.fit_transform(text)
print(vectorizer.get_feature_names_out())
print(X.toarray())
```
